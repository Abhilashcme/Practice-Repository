# CREATE TESTING AND TRAINING DATASET/DATA CLEANING

```python
# FEATURE SCALING/NORMALIZING THE 'AMOUNT' FEATURES

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df['Amount_Norm'] = sc.fit_transform(df['Amount'].values.reshape(-1,1))
```

```python
df
```

```python
df = df.drop(['Amount'], axis = 1)  # HERE WE'RE DROPPING THE 'AMOUNT' FEATURE
```

```python
# DROPING THE TARGET FEATURE i.e. 'CLASS' BEFORE TRAINING 

X = df.drop(['Class'],axis=1)
y = df['Class']
```

```python
X  #CLASS FEATURE IS DROPPED FROM THIS X
```

```python
y  #THAT DROPPED 'CLASS' FEATURE IS STORED IN y
```

```python
# SPLITING THE DATASET INTO TRAIN AND TEST USING train_test_split

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# TRAINING THE MODEL

```python
print("Shape of X_train: ",X_train.shape) 
print("Shape of y_train: ",y_train.shape)
print("Shape of X_test: ",X_test.shape)
print("Shape of y_test: ",y_test.shape)
```

```python
# TRAINING USING GAUSSIAN NAIVE BAYES
from sklearn.naive_bayes import GaussianNB 
NB_classifier = GaussianNB()
NB_classifier.fit(X_train, y_train)
```

#  EVALUATING THE MODEL 

```python
# PREDECTING ON THE TEST DATSET 

from sklearn.metrics import classification_report, confusion_matrix
y_predict = NB_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

## Unbalanced Dataset
## Since we are facing the issue of Unbalanced Dataset.
## Let's balance this dataset with techniques like OVERSAMPLING AND UNDERSAMPLING using SMOTE

## We'll basically UNDERSAMPLE THE MAJORITY CLASS DATASET as well as OVERSAMPLE THE MINORITY CLASS DATASET THAT WILL EVENTUALLY LEAD TO BALANCED DATASET

```python
import os
from imblearn.over_sampling import SMOTE  # Synthetic Minority Oversampling Technique

smote_X = X_train.copy()
smote_Y = y_train.copy()

os = SMOTE(random_state=0,k_neighbors=2)
os_smote_X, os_smote_Y = os.fit_sample(smote_X, smote_Y)
```

```python
NB_classifier = GaussianNB()
NB_classifier.fit(os_smote_X, os_smote_Y)
y_predict = NB_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

```python
# Handle class imbalance

# HERE WE ARE DOING UNDER SAMPLING
from imblearn.under_sampling import RandomUnderSampler

ros = RandomUnderSampler()
X_ros, y_ros = ros.fit_sample(X_train, y_train)

y_ros.value_counts()
```

```python
NB_classifier = GaussianNB()
NB_classifier.fit(X_ros, y_ros)
y_predict = NB_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

```python
# Handle class imbalance

# HERE WE'RE DOING OVERSAMPLING
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler()
X_ros, y_ros = ros.fit_sample(X_train, y_train)

y_ros.value_counts()
```
