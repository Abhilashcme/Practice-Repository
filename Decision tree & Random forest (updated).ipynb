```python
import pandas as pd
import numpy as np
```

```python
#reading data
employee=pd.read_csv('Repeat.csv')
```

```python
#printing first five rows
employee.head()
```

```python
#shape of the data - We have 1470 rows and 35 columns
employee.shape
```

```python
employee.describe()
```

```python
#checking for data types and null values
employee.info()
```

```python
#dropping the unnecessary columns
employee = employee.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1)
```

```python
#encoding attribute variable 
employee['Attrition']=np.where(employee['Attrition']=='Yes',1,0)
```

```python
employee.head()
```

# Visualizations

```python
import seaborn as sns
```

```python
#distribution of Attrition feature

sns.countplot("Attrition",data=employee)
```

```python
#attrition of employees based on the age

import matplotlib.pyplot as plt

plt.figure(figsize=(12,7))
sns.countplot('Age',hue='Attrition',data=employee)
```

```python
#attrition of employees based on department

sns.countplot('Department',hue='Attrition',data=employee)
```

```python
#monthly salary of employees based on their job roles

sns.barplot(employee['JobRole'],employee['MonthlyIncome'])
plt.xticks(rotation=90);
```

```python
#checking the values in categorical features

print(employee.BusinessTravel.value_counts())
print('***************************************')
print(employee.Department.value_counts())
print('***************************************')
print(employee.EducationField.value_counts())
print('***************************************')
print(employee.JobRole.value_counts())
print('***************************************')
```

```python
#encoding categorical variables into numerical variables
employee_dummies=pd.get_dummies(employee[['BusinessTravel','Department','EducationField','Gender',\
                                          'JobRole','MaritalStatus','OverTime']],drop_first=True)
```

```python
employee=employee.drop(['BusinessTravel','Department','EducationField','Gender',
                        'JobRole','MaritalStatus','OverTime'],axis=1)
```

```python
employee=pd.concat([employee,employee_dummies],axis=1)
```

```python
employee.head()
```

```python
train_x=employee.drop(['Attrition'],axis=1)
train_y=employee['Attrition']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_x, train_y, test_size=0.3, random_state=20)
```

## Building the Decision tree basic model

```python
#Import Tree Classifier model
from sklearn import tree
from sklearn import metrics


dt = tree.DecisionTreeClassifier() #creating decision tree object
#Train the model using the training sets
dt.fit(X_train, y_train)
```

```python
#Predicting on train data
x_pred=dt.predict(X_train)
```

```python
#checking the accuracy of train data

print ('Accuracy of train data :',metrics.accuracy_score(y_train,x_pred))

print("***************************************************")

print("Confusion matrix of train data :\n",metrics.confusion_matrix(y_train,x_pred))
```

```python
#Predict the response for test dataset
y_pred = dt.predict(X_test)
```

```python
#import scikit learn metrics to calculate accuracy of test data
from sklearn import metrics
print ('Accuracy:',metrics.accuracy_score(y_test,y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred)) # recall means how much % of your data is correctly depicted by our model

print("***************************************************")

print("Confusion matrix :\n",metrics.confusion_matrix(y_test,y_pred))
```

```python
#classification report of test data

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
#We can see that we are able to predict with 80 % accurecy will the person stay in the company or leave.
```

## Building Decision Tree model by specifying parameters

```python
#Import Tree Classifier model
from sklearn import tree

dt1 = tree.DecisionTreeClassifier(max_depth=5,min_samples_split=3) #creating decision tree object
#Train the model using the training sets
dt1.fit(X_train, y_train)
```

```python
#Predicting on train data
x_pred=dt1.predict(X_train)
```

```python
#checking the accuracy of train data

print ('Accuracy of train data :',metrics.accuracy_score(y_train,x_pred))

print("Confusion matrix of train data :\n",metrics.confusion_matrix(y_train,x_pred))

#as we did pre pruning of the tree overfitting started to reduce , as we can see that previously accuracy score was 1 and now
#new accuracy score is 0.9
```

```python
#Predict the response for test dataset
y_pred = dt1.predict(X_test)
```

```python
#import scikit learn metrics to calculate accuracy of test data
from sklearn import metrics
print ('Accuracy:',metrics.accuracy_score(y_test,y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))# recall means how much % of your data is correctly depicted by our model
print("Kappa score:",metrics.cohen_kappa_score(y_test, y_pred))

print("***************************************************")

print("Confusion matrix :\n",metrics.confusion_matrix(y_test,y_pred))
```

```python
# AUC and ROC curves

y_pred_proba = dt1.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="auc="+str(auc))
plt.plot([0, 1], [0, 1], color='r', linestyle='--')
plt.legend(loc=4)
plt.show()
```

## Grid search for Decision tree

```python
from sklearn.model_selection import GridSearchCV

# use a full grid over all parameters
param_grid = {"max_depth": [3, None],
              "max_features": [1, 3, 10],
              "min_samples_split": [2, 3, 10],
              "criterion": ["gini", "entropy"]}

# run grid search
grid_search = GridSearchCV(dt, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
# build a classifier with best of Grid search
dt_grid_search = tree.DecisionTreeClassifier(criterion = 'entropy',  max_depth=3, max_features = 10, min_samples_split = 10)

#Train the model using the training sets with best of Grid search
dt_grid_search.fit(X_train, y_train)

#Predict the response for test dataset with best of Grid search
y_pred = dt_grid_search.predict(X_test)

# Model Accuracy , Precision, Recall
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

# Ensemble - Random Forest

```python
#import random forest 
from sklearn.ensemble import RandomForestClassifier

# build a classifier
rf = RandomForestClassifier(n_estimators=100)

#Train the model using the training sets
rf.fit(X_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = rf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
# Model Precision
print("Precision:",metrics.precision_score(y_test, y_pred))
# Model Recall
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
# AUC and ROC curves

y_pred_proba = rf.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="auc="+str(auc))
plt.plot([0, 1], [0, 1], color='r', linestyle='--')
plt.legend(loc=4)
plt.show()
```

## GRID SEARCH

```python
from sklearn.model_selection import GridSearchCV

# use a full grid over all parameters
param_grid = {"max_depth": [3, None],
              "max_features": [1, 3, 10],
              "min_samples_split": [2, 3, 10],
              "bootstrap": [True, False],
              "criterion": ["gini", "entropy"]}

# run grid search
grid_search = GridSearchCV(rf, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
# build a classifier with best of Grid search
rf_grid_search = RandomForestClassifier(n_estimators=100, bootstrap = False , criterion = 'entropy',  max_features = 10, min_samples_split = 3)

#Train the model using the training sets with best of Grid search
rf_grid_search.fit(X_train, y_train)

#Predict the response for test dataset with best of Grid search
y_pred = rf_grid_search.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
# Model Precision
print("Precision:",metrics.precision_score(y_test, y_pred))
# Model Recall
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
# AUC and ROC curves

y_pred_proba = rf_grid_search.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="auc="+str(auc))
plt.plot([0, 1], [0, 1], color='r', linestyle='--')
plt.legend(loc=4)
plt.show()
```
