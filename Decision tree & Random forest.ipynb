```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',500)
```

```python
#reading data
employee=pd.read_csv('employee.csv')
```

```python
print (employee.EmployeeCount.unique())
print (employee.EmployeeNumber.unique())
print (employee.Over18.unique())
print (employee.StandardHours.unique())
```

```python
#dropping the unnecessary columns
employee = employee.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1)
```

```python
#encoding attribute variable 
employee['Attrition']=np.where(employee['Attrition']=='Yes',1,0)
```

```python
employee.head()
```

```python
#checking the values in categorical features

print(employee.BusinessTravel.value_counts())
print('***************************************')
print(employee.Department.value_counts())
print('***************************************')
print(employee.EducationField.value_counts())
print('***************************************')
print(employee.JobRole.value_counts())
print('***************************************')
```

```python
import seaborn as sns

# distribution of Attribute
sns.countplot("Attrition",data=employee)
```

```python
#attrition of employees based on the age

import matplotlib.pyplot as plt

plt.figure(figsize=(12,7))
sns.countplot('Age',hue='Attrition',data=employee)
```

```python
#attrition of employees based on department

sns.countplot('Department',hue='Attrition',data=employee)
```

```python
#encoding categorical variables into numerical variables
employee_dummies=pd.get_dummies(employee[['BusinessTravel','Department','EducationField','Gender',\
                                          'JobRole','MaritalStatus','OverTime']],drop_first=True)
```

```python
employee=employee.drop(['BusinessTravel','Department','EducationField','Gender',
                        'JobRole','MaritalStatus','OverTime'],axis=1)
```

```python
employee=pd.concat([employee,employee_dummies],axis=1)
```

```python
employee.head()
```

```python
train_x=employee.drop(['Attrition'],axis=1)
train_y=employee['Attrition']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_x, train_y, test_size=0.3, random_state=20)
```

## Building the Decision Tree Basic Model

```python
#Import Tree Classifier model
from sklearn import tree
from sklearn import metrics


dt = tree.DecisionTreeClassifier() #creating decision tree object
#Train the model using the training sets
dt.fit(X_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = dt.predict(X_train)
```

```python
#import scikit learn metrics to calculate accuracy
from sklearn import metrics
print ('Accuracy:',metrics.accuracy_score(y_test,y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred)) # recall means how much % of your data is correctly depicted by our model
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
#We can see that we are able to predict with 80 % accurecy will the person stay in the company or leave.
```

# Ensemble - Random Forest

```python
#import random forest 
from sklearn.ensemble import RandomForestClassifier

# build a classifier
rf = RandomForestClassifier(n_estimators=100)

#Train the model using the training sets
rf.fit(X_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = rf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
# Model Precision
print("Precision:",metrics.precision_score(y_test, y_pred))
# Model Recall
print("Recall:",metrics.recall_score(y_test, y_pred))
```

## GRID SEARCH

```python
from sklearn.model_selection import GridSearchCV

# use a full grid over all parameters
param_grid = {"max_depth": [3, None],
              "max_features": [1, 3, 10],
              "min_samples_split": [2, 3, 10],
              "bootstrap": [True, False],
              "criterion": ["gini", "entropy"]}

# run grid search
grid_search = GridSearchCV(rf, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
# build a classifier with best of Grid search
rf_grid_search = RandomForestClassifier(n_estimators=100, bootstrap = True , criterion = 'gini',  max_features = 10, min_samples_split = 2)

#Train the model using the training sets with best of Grid search
rf_grid_search.fit(X_train, y_train)

#Predict the response for test dataset with best of Grid search
y_pred = rf_grid_search.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
# Model Precision
print("Precision:",metrics.precision_score(y_test, y_pred))
# Model Recall
print("Recall:",metrics.recall_score(y_test, y_pred))
```
